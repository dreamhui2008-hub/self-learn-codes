### Chapter 2.7 - Documentation

Documentation is written information that explains code objects: what they do, what inputs they expect, and what outputs they return.

Learning to inspect documentation is part of learning ML because frameworks are too large to memorize.


```python
import inspect
import torch
```

### 2.7.1 Functions and Classes in a Module

#### 1. Intuition

A module is a Python file or package that contains reusable code. When we write `import torch`, we load the PyTorch module.

A function is a reusable block of code that can receive inputs and return an output.

A class is a template for creating objects. An object is a concrete value created from a class. A method is a function attached to an object or class.

#### 2. Why this exists

Frameworks contain thousands of functions and classes. Documentation tools help you discover what exists without memorizing everything.


#### 3. Examples

Use `dir` to list names inside a module. The result can be large, so we filter it.


```python
names = dir(torch)
rand_names = [name for name in names if name.startswith("rand")]
rand_names[:5]
```

Create a tiny class to make object terminology concrete.


```python
class Counter:
    def __init__(self, start=0):
        self.value = start
    def add(self, amount=1):
        self.value += amount
        return self.value

counter = Counter(2)
counter.add()
```

Inspect names attached to the object.


```python
public_names = [name for name in dir(counter) if not name.startswith("_")]
public_names
```

#### 4. Step-by-step breakdown

`dir(torch)` returns names available inside the `torch` module.

The list comprehension filters names that start with `rand`.

`class Counter:` defines a class named `Counter`.

`__init__` is a special method Python calls when creating a new object.

`self` is the object currently being created or used. In `self.value = start`, Python stores an attribute named `value` on that object.

An attribute is a named piece of data attached to an object.

`counter = Counter(2)` creates an object from the class.

`counter.add()` calls the `add` method on that object.

#### 5. Connection to ML systems

PyTorch uses modules, functions, classes, objects, methods, and attributes everywhere. For example, neural network layers are objects, and model parameters are attributes stored inside those objects.

#### 6. Common confusion points

- A module groups code.
- A class is a template; an object is an instance created from that template.
- `self` is not a magic global variable. It is the object receiving the method call.
- `dir` shows names, but it does not explain them deeply.


### 2.7.2 Specific Functions and Classes

#### 1. Intuition

To understand a specific function or class, look for its signature, docstring, parameters, return value, and examples.

A signature is the visible pattern of inputs a function accepts. A parameter is a named input in a function definition. An argument is the actual value passed when calling the function.

A docstring is text written inside code to explain what it does.

#### 2. Why this exists

When code fails, the answer is often in the function documentation: wrong argument name, wrong dtype, wrong shape, or misunderstood return value.


#### 3. Examples

Define a small documented function.


```python
def scale(x, factor=2):
    """Multiply x by a chosen factor."""
    return x * factor

scale(3), scale(3, factor=10)
```

Inspect the function signature and docstring.


```python
signature = inspect.signature(scale)
doc = scale.__doc__
signature, doc
```

Inspect a PyTorch function docstring. The full text can be long, so take the first few lines.


```python
doc = torch.ones.__doc__
first_lines = doc.split("\n")[:5]
first_lines
```

Use `help` when you want Python to print documentation in a readable format.


```python
help(scale)
```

#### 4. Step-by-step breakdown

`factor=2` gives the parameter `factor` a default value. A default value is used when the caller does not provide that argument.

`scale(3)` uses the default factor of 2.

`scale(3, factor=10)` overrides the default.

`inspect.signature(scale)` returns the input pattern for the function.

`scale.__doc__` reads the function's docstring.

`torch.ones.__doc__` reads PyTorch documentation stored on the function object.

`help(scale)` asks Python to display help information.

#### 5. Connection to ML systems

Research code often uses unfamiliar framework calls. Instead of guessing, inspect the function signature and documentation, then test with a tiny example.

#### 6. Common confusion points

- Parameter and argument are related but not identical words.
- Defaults are used only when an argument is omitted.
- Documentation may assume background knowledge, so tiny experiments are still useful.
- Some framework functions are implemented in lower-level code, so `inspect.signature` may not always work on them.
