# Chapter 01 - Functions as Real Objects

#### 1. Mechanical Refresher
In Python, a function is an object: you can store it in a variable, pass it into another function, return it from a function, and call it later with `()`. A closure is an inner function that keeps access to variables from the outer function call that created it, and `*args` / `**kwargs` collect or unpack positional and keyword arguments when you need a forwarding function.

#### 2. Minimal Working Example

```python
def twice(x):
    return x * 2

def apply(func, value):
    return func(value)

def make_offset(delta):
    def offset(x):
        return x + delta
    return offset

add_three = make_offset(3)
print(apply(twice, 5))
print(add_three(10))
```

`apply` receives a function object and calls it later. `make_offset` returns the inner `offset` function instead of calling it immediately. `offset` still remembers `delta` after `make_offset` has finished.

#### 8. Common Bugs
- Calling too early: `func = square(4)` stores the result, not the function; the symptom is usually `TypeError: 'int' object is not callable`.
- Forgetting `return` for an inner function: the outer factory returns `None`; the symptom appears later when you try to call the returned value.
- Forgetting `*` while forwarding arguments: the callee receives one tuple instead of several arguments; the symptom is a wrong argument count or wrong output.
- Mutating captured data without meaning to: a closure can keep a list or dictionary alive; the symptom is state from a previous call appearing in a later call.
