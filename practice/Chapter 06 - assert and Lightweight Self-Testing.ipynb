# Chapter 06 - `assert` and Lightweight Self-Testing

#### 1. Mechanical Refresher
`assert condition, message` checks that `condition` is true at that line; if it is false, Python raises an `AssertionError` with the optional message. In production-style practice, asserts are best for programmer-facing invariants and sanity checks inside code or small test functions, not for user-facing validation.

#### 2. Minimal Working Example

```python
def normalize_score(score, total):
    assert total > 0, "total must be positive"
    assert 0 <= score <= total, "score outside range"
    return score / total

value = normalize_score(3, 4)
assert value == 0.75
print("checks passed")
```

The first two asserts protect the assumptions required by the calculation. The final assert is a tiny test: it checks the actual return value against a known expected value.

```python
def average(values):
    assert len(values) > 0, "values must not be empty"
    total = 0
    for value in values:
        total = total + value
    return total / len(values)

actual = average([2, 4, 6])
assert actual == 4
print("passed:", actual)
```

#### 7. Applied AI/ML Drill
**ML to Python mirror:** tensor-shape sanity checks are ordinary asserts about list lengths before a calculation depends on them. **Python to ML mirror:** real model code often asserts or checks batch size, feature count, and target length before a forward pass so shape bugs fail at the boundary instead of producing confusing downstream results.

**Applied Drill.** Add shape asserts before a tiny linear forward pass. Expected behavior: predictions `[6, 14]`.

```python
def linear_forward(batch, weights):
    assert len(batch) > 0, "batch must not be empty"
    predictions = []
    for row in batch:
        # TODO: assert row length matches weights length.
        total = 0
        for value, weight in zip(row, weights):
            total = total + value * weight
        predictions.append(total)
    return predictions

actual = linear_forward([[1, 2], [3, 4]], [2, 2])
assert actual == [6, 14]
print("passed:", actual)
```

#### 8. Common Bugs
- Using assert for user input validation: asserts can be disabled with optimization flags, so they are for developer assumptions.
- Placing the assert after the unsafe operation: the symptom is a lower-level error before your clearer message can run.
- Writing `assert actual = expected`: assignment is not comparison; use `==`.
- Asserting a broad truth without a message: the symptom is an unhelpful `AssertionError` when the check fails.

#### 9. Compounding Drill

Combine asserts with Chapter 5 generators: verify batch sizes while lazily yielding chunks.

```python
def checked_batches(values, batch_size):
    assert batch_size > 0, "batch_size must be positive"
    index = 0
    while index < len(values):
        batch = values[index:index + batch_size]
        # TODO: assert the yielded batch is not empty.
        yield batch
        index = index + batch_size

actual = []
for batch in checked_batches([1, 2, 3], 2):
    actual.append(batch)
assert actual == [[1, 2], [3]]
print("passed:", actual)
```

```python
# Compounding Drill and Chapter Project approach
def checked_batches(values, batch_size):
    assert batch_size > 0, "batch_size must be positive"
    index = 0
    while index < len(values):
        batch = values[index:index + batch_size]
        assert len(batch) > 0, "empty batch should never be yielded"
        yield batch
        index = index + batch_size

assert list(checked_batches([1, 2, 3], 2)) == [[1, 2], [3]]

def log_batch_size(func):
    def wrapper(*args, **kwargs):
        for features, targets in func(*args, **kwargs):
            print("batch size:", len(features))
            yield features, targets
    return wrapper

@log_batch_size
def checked_minibatches(features, targets, batch_size):
    assert batch_size > 0, "batch_size must be positive"
    assert len(features) == len(targets), "features and targets must align"
    index = 0
    while index < len(features):
        yield features[index:index + batch_size], targets[index:index + batch_size]
        index = index + batch_size

assert list(checked_minibatches([[1], [2], [3]], [2, 4, 6], 2)) == [([[1], [2]], [2, 4]), ([[3]], [6])]
print("project solution passed")
```
