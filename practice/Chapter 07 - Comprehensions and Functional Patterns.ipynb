# Chapter 07 - Comprehensions and Functional Patterns

#### 1. Mechanical Refresher
A comprehension builds a list, dictionary, set, or lazy generator expression by putting a loop-shaped expression inside brackets, braces, or parentheses. `map`, `filter`, `functools.reduce`, and `functools.partial` are function-based tools for transforming, selecting, combining, or pre-filling arguments, but they should be used only when they make the code easier to read than a plain loop or comprehension.

#### 2. Minimal Working Example

```python
numbers = [1, 2, 3, 4]
squares = [n * n for n in numbers]
by_value = {n: n * n for n in numbers}
evens = {n for n in numbers if n % 2 == 0}
lazy = (n * n for n in numbers)

assert squares == [1, 4, 9, 16]
assert by_value == {1: 1, 2: 4, 3: 9, 4: 16}
assert evens == {2, 4}
assert next(lazy) == 1
print("checks passed")
```

The list comprehension stores all square values immediately. The dictionary and set comprehensions choose their container shape. The generator expression stores the loop recipe and produces one value when `next` asks for it.

```python
from functools import partial

def scale(value, factor=1):
    return value * factor

triple = partial(scale, factor=3)
actual = [triple(n) for n in [2, 4]]
expected = [6, 12]
assert actual == expected
print("passed:", actual)
```

#### 8. Common Bugs
- Reversing expression and loop order: the symptom is `SyntaxError` or a comprehension that reads backward.
- Using a comprehension for too much nested logic: the symptom is code you cannot quickly predict; use a loop when readability drops.
- Forgetting to materialize `map` or `filter`: the symptom is printing a map/filter object instead of values.
- Omitting the initializer in `reduce`: the symptom appears on empty inputs or when the first value is not a safe starting accumulator.
