### Chapter 2.3 - Linear Algebra

Linear algebra is the language of structured numerical computation. It gives names to numbers, lists of numbers, tables of numbers, and operations that combine them.

Deep learning uses linear algebra because model inputs, weights, activations, gradients, and losses are all built from tensors.


```python
import numpy as np
import torch
```

### 2.3.1 Scalars

#### 1. Intuition

A scalar is a single number. Examples include `3`, `-1.5`, and `0`.

In ML, a loss value is often a scalar. A learning rate is also a scalar. A learning rate is a small number that controls how large a model update should be.

#### 2. Why this exists

Scalars are the simplest numerical objects. More complex objects, such as vectors and matrices, are built from scalars.


#### 3. Examples

Plain Python uses ordinary numbers as scalars.


```python
x = 3.0
y = 2.0
x + y, x * y, x / y
```

PyTorch can store a scalar as a 0-dimensional tensor.


```python
x = torch.tensor(3.0)
y = torch.tensor(2.0)
z = x * y
z.shape, z
```

#### 4. Step-by-step breakdown

`x = 3.0` stores a Python floating-point number.

`torch.tensor(3.0)` stores one number in a PyTorch tensor.

`z.shape` is empty for a scalar tensor because there are no axes to move along.

A scalar tensor and a Python scalar can display similarly, but they are different object types.

#### 5. Connection to ML systems

Scalar losses are important because PyTorch's default `.backward()` call expects a scalar output. Backward means asking PyTorch to compute gradients for earlier tensors.

#### 6. Common confusion points

- A scalar is one number, not a list with one number.
- Shape `()` means scalar tensor.
- Shape `(1,)` means one-dimensional tensor with one element.
- A scalar can still be part of a larger computation graph in PyTorch.


### 2.3.2 Vectors

#### 1. Intuition

A vector is an ordered list of numbers. Ordered means position matters.

A feature vector stores several measurements for one example. For a house, a vector might store `[size, bedrooms, age]`.

#### 2. Why this exists

Vectors let us treat several related numbers as one object. This is useful when one example has multiple features.


#### 3. Examples

Plain Python can represent a vector as a list.


```python
x = [1.0, 2.0, 3.0]
length = len(x)
second = x[1]
length, second
```

NumPy and PyTorch store vectors with shape information.


```python
x_np = np.array([1.0, 2.0, 3.0])
x_torch = torch.tensor([1.0, 2.0, 3.0])
x_np.shape, x_torch.shape
```

#### 4. Step-by-step breakdown

`len(x)` counts the number of entries in the Python list.

`x[1]` selects the second value because indexing starts at 0.

`np.array([...])` creates a NumPy vector.

`torch.tensor([...])` creates a PyTorch vector.

Both vector shapes are `(3,)`, meaning one dimension with three values.

#### 5. Connection to ML systems

A single training example is often represented as a vector. A batch of examples becomes a matrix, where each row is one vector.

#### 6. Common confusion points

- Vector length is the number of entries.
- A row vector and column vector may have different shapes.
- Shape `(3,)` is not the same as `(3, 1)`.
- Position matters: swapping entries changes the meaning.


### 2.3.3 Matrices

#### 1. Intuition

A matrix is a rectangular table of numbers with rows and columns.

A batch of tabular examples is often a matrix. Rows represent examples. Columns represent features.

#### 2. Why this exists

Matrices let us process many vectors together. This is one reason deep learning code can be fast: it groups many examples into one matrix operation.


#### 3. Examples

Plain Python can represent a matrix as a list of lists.


```python
A = [[1, 2, 3], [4, 5, 6]]
rows = len(A)
cols = len(A[0])
A[1][2], (rows, cols)
```

PyTorch gives direct shape and transpose operations. Transpose means swapping rows and columns.


```python
A = torch.tensor([[1, 2, 3], [4, 5, 6]])
A.shape, A.T.shape
```

#### 4. Step-by-step breakdown

`A[1][2]` first selects row 1, then column 2.

`A.shape` returns `(2, 3)`, meaning 2 rows and 3 columns.

`A.T` returns the transpose. The original shape `(2, 3)` becomes `(3, 2)`.

The transpose does not change the numbers themselves. It changes their arrangement.

#### 5. Connection to ML systems

Weight matrices are central to neural networks. A linear layer often multiplies an input matrix by a weight matrix to produce an output matrix.

#### 6. Common confusion points

- Rows and columns are both dimensions.
- Transpose swaps dimensions for a matrix.
- A matrix is 2-dimensional, even if one dimension has size 1.
- A batch matrix's row meaning is a convention chosen by the code.


### 2.3.4 Tensors

#### 1. Intuition

In linear algebra for deep learning, tensor is the general word for a numerical array with any number of dimensions.

A scalar is a 0-dimensional tensor. A vector is a 1-dimensional tensor. A matrix is a 2-dimensional tensor. A stack of matrices is a 3-dimensional tensor.

#### 2. Why this exists

Real ML data often needs more than two dimensions. An image batch may use dimensions for examples, color channels, height, and width.


#### 3. Examples

Create a tiny 3-dimensional tensor.


```python
T = torch.arange(8).reshape(2, 2, 2)
T.shape
```

Indexing a higher-dimensional tensor removes or selects dimensions.


```python
first_block = T[0]
one_value = T[1, 0, 1]
first_block.shape, one_value
```

#### 4. Step-by-step breakdown

`torch.arange(8)` creates values 0 through 7.

`reshape(2, 2, 2)` arranges those 8 values into three dimensions.

`T[0]` selects the first block, leaving a 2 by 2 matrix.

`T[1, 0, 1]` selects one scalar value by giving an index for every dimension.

#### 5. Connection to ML systems

Image, audio, and text models all rely on tensors with more than two dimensions. The shape records the organization that gives the numbers meaning.

#### 6. Common confusion points

- In PyTorch, tensor is the general container, not only a high-dimensional object.
- More dimensions do not mean more abstract math. They usually mean more indexing directions.
- Shape order matters. `(batch, height, width)` is not the same as `(height, width, batch)`.
- Indexing can reduce the number of dimensions.


### 2.3.5 Basic Properties of Tensor Arithmetic

#### 1. Intuition

Tensor arithmetic follows familiar arithmetic rules when operations are valid.

Elementwise arithmetic applies the operation separately at each matching position.

#### 2. Why this exists

These properties let us simplify expressions and reason about tensor code. However, shape compatibility must still be checked.


#### 3. Examples

Check commutativity for elementwise addition. Commutativity means order does not change the result.


```python
x = torch.tensor([1, 2, 3])
y = torch.tensor([10, 20, 30])
x + y, y + x
```

Check distributivity. Distributivity means multiplication can be spread across addition.


```python
a = torch.tensor([2, 2, 2])
x = torch.tensor([1, 2, 3])
y = torch.tensor([4, 5, 6])
a * (x + y), a * x + a * y
```

#### 4. Step-by-step breakdown

`x + y` and `y + x` match because addition is commutative.

`a * (x + y)` first adds `x` and `y`, then multiplies by `a`.

`a * x + a * y` first multiplies separately, then adds.

The two results match for elementwise operations with compatible shapes.

#### 5. Connection to ML systems

Loss functions and model layers are made of tensor expressions. Algebraic reasoning helps debug whether two implementations should produce the same values.

#### 6. Common confusion points

- Familiar arithmetic rules apply only when shapes and operations are compatible.
- Elementwise multiplication is not matrix multiplication.
- Broadcasting can make expressions valid, but it can also hide shape mistakes.
- Equality of displayed values does not mean two tensors share memory.


### 2.3.6 Reduction

#### 1. Intuition

A reduction operation turns many values into fewer values. Summing all entries of a vector reduces it to one scalar.

Common reductions include sum, mean, maximum, and minimum.

#### 2. Why this exists

ML often needs to summarize many values. A loss over many examples may be averaged into one scalar. Accuracy may count many correct predictions and divide by total examples.


#### 3. Examples

Manual Python sum over a vector.


```python
x = [1, 2, 3]
total = 0
for value in x:
    total += value
total
```

PyTorch reductions can reduce all values or reduce along one dimension.


```python
A = torch.tensor([[1, 2, 3], [4, 5, 6]])
all_sum = A.sum()
col_sum = A.sum(dim=0)
row_sum = A.sum(dim=1)
all_sum, col_sum, row_sum
```

#### 4. Step-by-step breakdown

The manual loop starts with `total = 0` and adds each value one by one.

`A.sum()` reduces every element into one scalar.

`A.sum(dim=0)` reduces along rows, so it returns one value per column.

`A.sum(dim=1)` reduces along columns, so it returns one value per row.

#### 5. Connection to ML systems

Training often reduces per-example losses into one scalar loss. That scalar is then used for gradient computation.

#### 6. Common confusion points

- Reduction changes shape.
- `dim=0` means reduce over the first dimension.
- `mean` is a reduction too.
- A scalar result may have shape `()`, not `(1,)`.


### 2.3.7 Non-Reduction Sum

#### 1. Intuition

A non-reduction sum keeps the reduced dimension as size 1. It summarizes values but preserves enough shape structure for later broadcasting.

In PyTorch, this is done with `keepdim=True`.

#### 2. Why this exists

Keeping dimensions can prevent shape errors. It is especially useful when normalizing rows or columns.


#### 3. Examples

Compare reduction with and without keeping dimensions.


```python
A = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
row_sum = A.sum(dim=1)
row_sum_keep = A.sum(dim=1, keepdim=True)
row_sum.shape, row_sum_keep.shape
```

Use the kept dimension for broadcasting.


```python
A = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
row_sum = A.sum(dim=1, keepdim=True)
row_fraction = A / row_sum
row_fraction
```

#### 4. Step-by-step breakdown

`A.sum(dim=1)` returns shape `(2,)` because the column dimension is removed.

`A.sum(dim=1, keepdim=True)` returns shape `(2, 1)` because the reduced dimension is kept with size 1.

`A / row_sum` divides each row by its own row sum using broadcasting.

#### 5. Connection to ML systems

Normalization often uses this pattern. Normalization means scaling values into a more controlled range or relationship, such as making each row sum to 1.

#### 6. Common confusion points

- `keepdim=True` does not keep the original values; it keeps the dimension position.
- Shape `(2,)` and `(2, 1)` behave differently in broadcasting.
- Non-reduction sum is still a summary operation.
- The name is about shape preservation, not about avoiding addition.


### 2.3.8 Dot Products

#### 1. Intuition

A dot product combines two equal-length vectors into one scalar. It multiplies matching entries and then sums the products.

For vectors `x = [x1, x2]` and `y = [y1, y2]`, the dot product is `x1*y1 + x2*y2`.

#### 2. Why this exists

Dot products measure weighted combinations and similarity. If two vectors put large values in the same positions, their dot product tends to be large.


#### 3. Examples

Manual Python dot product.


```python
x = [1, 2, 3]
y = [10, 20, 30]
products = [x[i] * y[i] for i in range(len(x))]
dot = sum(products)
dot
```

PyTorch dot product.


```python
x = torch.tensor([1.0, 2.0, 3.0])
y = torch.tensor([10.0, 20.0, 30.0])
torch.dot(x, y)
```

#### 4. Step-by-step breakdown

The manual code loops over positions.

At each position, it multiplies one value from `x` by the matching value from `y`.

`sum(products)` reduces the product vector into one scalar.

`torch.dot(x, y)` performs the same operation for 1-dimensional tensors.

#### 5. Connection to ML systems

Linear regression uses dot products. A prediction can be computed as a dot product between feature values and model weights, plus a bias.

#### 6. Common confusion points

- Dot product output is a scalar.
- The two vectors must have the same length.
- Dot product is not elementwise multiplication alone; it includes a final sum.
- `torch.dot` expects 1-dimensional tensors.


### 2.3.9 Matrix-Vector Products

#### 1. Intuition

A matrix-vector product multiplies a matrix by a vector and returns a vector.

You can think of each row of the matrix taking a dot product with the vector.

#### 2. Why this exists

This operation lets one vector be transformed into another vector. Neural network layers use this idea constantly.


#### 3. Examples

Manual matrix-vector product.


```python
A = [[1, 2], [3, 4]]
x = [10, 20]
y = []
for row in A:
    y.append(sum(row[i] * x[i] for i in range(2)))
y
```

PyTorch matrix-vector product.


```python
A = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
x = torch.tensor([10.0, 20.0])
y = torch.mv(A, x)
y
```

#### 4. Step-by-step breakdown

The manual loop visits each row of `A`.

For each row, it computes a dot product with `x`.

The first output value comes from the first row. The second output value comes from the second row.

`torch.mv(A, x)` means matrix-vector multiply.

#### 5. Connection to ML systems

A linear layer can be understood as applying a matrix-vector product to one example. With batches, this becomes matrix-matrix multiplication.

#### 6. Common confusion points

- The vector length must match the number of matrix columns.
- The output length equals the number of matrix rows.
- Matrix-vector multiplication is not elementwise multiplication.
- Row meaning and column meaning must be tracked carefully.


### 2.3.10 Matrix-Matrix Multiplication

#### 1. Intuition

Matrix-matrix multiplication combines two matrices. Each output entry is a dot product between one row from the left matrix and one column from the right matrix.

#### 2. Why this exists

This operation processes many vectors at once. Batches, linear layers, attention mechanisms, and many scientific computations use matrix multiplication heavily.


#### 3. Examples

Manual matrix-matrix multiplication for tiny matrices.


```python
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]
C = []
for i in range(2):
    row = []
    for j in range(2):
        row.append(A[i][0] * B[0][j] + A[i][1] * B[1][j])
    C.append(row)
C
```

PyTorch uses `@` or `torch.mm` for matrix multiplication.


```python
A = torch.tensor([[1.0, 2.0], [3.0, 4.0]])
B = torch.tensor([[5.0, 6.0], [7.0, 8.0]])
A @ B, torch.mm(A, B)
```

#### 4. Step-by-step breakdown

The manual code builds one output row at a time.

For output position `(i, j)`, it combines row `i` of `A` with column `j` of `B`.

`A @ B` is Python's matrix multiplication operator.

`torch.mm(A, B)` is PyTorch's matrix-matrix multiplication function.

#### 5. Connection to ML systems

Batch training often computes many predictions using one matrix multiplication. This is much faster than predicting examples one by one in Python loops.

#### 6. Common confusion points

- For `A @ B`, the columns of `A` must match the rows of `B`.
- The output shape is `(rows of A, columns of B)`.
- `@` is not the same as `*`.
- Matrix multiplication order matters: `A @ B` may differ from `B @ A`.


### 2.3.11 Norms

#### 1. Intuition

A norm is a function that measures the size or length of a vector or matrix.

For a vector, the L2 norm is the ordinary Euclidean length. The L1 norm is the sum of absolute values.

#### 2. Why this exists

Norms help measure magnitude. In ML, norms are used in regularization, gradient clipping, distance calculations, and error measurement.

Regularization means adding a preference that discourages overly large or complex model parameters.


#### 3. Examples

Manual L1 and L2 norms for a vector.


```python
x = [3.0, 4.0]
l1 = sum(abs(v) for v in x)
l2 = sum(v * v for v in x) ** 0.5
l1, l2
```

PyTorch norm calculations.


```python
x = torch.tensor([3.0, 4.0])
l2 = torch.norm(x)
l1 = torch.abs(x).sum()
l1, l2
```

#### 4. Step-by-step breakdown

`abs(v)` gives the distance of `v` from zero, ignoring sign.

The L1 norm adds absolute values.

The L2 norm squares values, adds them, then takes the square root.

`torch.norm(x)` computes the default L2 norm for this vector.

`torch.abs(x).sum()` computes the L1 norm.

#### 5. Connection to ML systems

A large gradient norm may mean an unstable update. A large parameter norm may mean the model is using very large weights. Later chapters use norms to control these behaviors.

#### 6. Common confusion points

- Norms return nonnegative values.
- L1 and L2 norms are different size measures.
- A norm is not the same as a sum unless you specify L1 for nonnegative values.
- Norms are summaries, so they usually reduce many values into one value.


### 2.3.12 Discussion

#### 1. Intuition

Linear algebra gives a compact way to describe repeated numerical patterns.

Instead of writing many loops, we describe vectors, matrices, tensors, and operations between them.

#### 2. Why this exists

Deep learning frameworks are built around these operations because hardware can run them efficiently. GPUs are especially good at large matrix and tensor operations.


#### 3. Examples

A tiny linear prediction can be written as a dot product.


```python
features = torch.tensor([2.0, 3.0])
weights = torch.tensor([10.0, 1.0])
bias = torch.tensor(5.0)
prediction = torch.dot(features, weights) + bias
prediction
```

#### 4. Step-by-step breakdown

`features` stores one example.

`weights` stores how much each feature contributes.

`torch.dot(features, weights)` computes the weighted sum.

`bias` shifts the result by a scalar amount.

This is the core computation behind linear regression.

#### 5. Connection to ML systems

Later neural networks repeat and compose this idea. Layers transform tensors using parameter tensors, and training adjusts those parameters.

#### 6. Common confusion points

- Linear algebra notation hides loops, but the loops still conceptually exist.
- Shape tracking is part of understanding the math.
- A correct formula with wrong tensor shapes still fails.
- Efficient code often looks shorter because the library handles repeated work.
